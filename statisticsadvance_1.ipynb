##**THEORY QUESTIONS**

**1.What is a random variable in probability theory?**
- random variable is a numerical outcome of a random phenomenon, which can take on different values based on the outcome of a random event.

---



**2.What are the types of random variables?**
- There are two main types of random variables: discrete random variables, which take on a countable number of distinct values, and continuous random variables, which can take on any value within a given range.

---



**3.What is the difference between discrete and continuous distributions?**
- Discrete distributions describe random variables that can take on specific, separate values (e.g., integers), while continuous distributions describe random variables that can take on any value within a continuous range (e.g., real numbers).

---



**4.What are probability distribution functions (PDF)?**
- A probability distribution function (PDF) describes the likelihood of a random variable taking on a particular value. For discrete variables, it is called a probability mass function (PMF), while for continuous variables, it is a function that describes the density of probabilities over a range of values.

---



**5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
- A CDF gives the probability that a random variable is less than or equal to a certain value, while a PDF describes the likelihood of a random variable taking on a specific value. For continuous variables, the PDF is the derivative of the CDF.

---



**6.What is a discrete uniform distribution?**
- A discrete uniform distribution is a probability distribution where a finite number of outcomes are equally likely. For example, rolling a fair die has a discrete uniform distribution with outcomes {1, 2, 3, 4, 5, 6}.

---



**7.What are the key properties of a Bernoulli distribution?**
- Bernoulli distribution has two possible outcomes (success and failure), with a single parameter ( p ) representing the probability of success. The mean is ( p ) and the variance is ( p(1-p) ).

---



**8.What is the binomial distribution, and how is it used in probability?**
- The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success ( p ). It is used to calculate probabilities for scenarios like coin flips or quality control in manufacturing.

---



**9.What is the Poisson distribution and where is it applied?**
- The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence. It is often used in fields like telecommunications, traffic flow, and queuing theory.

---



**10.What is a continuous uniform distribution?**
- continuous uniform distribution is a probability distribution where all outcomes in a continuous range are equally likely. It is defined by two parameters, ( a ) and ( b ), which are the minimum and maximum values.

---



**11.What are the characteristics of a normal distribution?**
- Normal distribution is symmetric, bell-shaped, and characterized by its mean (µ) and standard deviation (σ). About 68% of the data falls within one standard deviation of the mean, 95% within two, and 99.7% within three.

---



**12.What is the standard normal distribution, and why is it important?**
- The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it allows for the standardization of scores (Z-scores), facilitating comparisons across different normal distributions.

---



**13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
- The Central Limit Theorem states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution of the population. It is critical because it justifies the use of normal distribution in inferential statistics.

---



**14.How does the Central Limit Theorem relate to the normal distribution?**
- The CLT indicates that as the sample size grows, the sampling distribution of the sample mean will approximate a normal distribution, allowing statisticians to make inferences about population parameters using normal distribution properties.

---



**15.What is the application of Z statistics in hypothesis testing?**
- Z statistics are used in hypothesis testing to determine how many standard deviations a sample mean is from the population mean under the null hypothesis. It helps assess whether to reject or fail to reject the null hypothesis based on the calculated Z-score and corresponding p-value.

---



**16.How do you calculate a Z-score, and what does it represent?**
- Z-score is calculated using the formula:
[ Z = \frac{(X - \mu)}{\sigma} ]
where ( X ) is the value, ( \mu ) is the mean, and ( \sigma ) is the standard deviation. It represents the number of standard deviations a data point is from the mean, indicating how unusual or typical the value is within the distribution.

---



**17.What are point estimates and interval estimates in statistics?**
- point estimate provides a single value as an estimate of a population parameter (e.g., sample mean as an estimate of population mean). An interval estimate provides a range of values (e.g., confidence interval) within which the parameter is expected to lie, reflecting uncertainty.

---



**18.What is the significance of confidence intervals in statistical analysis?**
- Confidence intervals provide a range of values that likely contain the true population parameter, offering a measure of uncertainty around point estimates. They help assess the reliability of estimates and make informed decisions based on data.

---



**19.What is the relationship between a Z-score and a confidence interval?**
- Z-score is used to determine the critical values that define the boundaries of a confidence interval. For example, a 95% confidence interval corresponds to Z-scores of approximately ±1.96, indicating the range within which the true parameter is expected to fall with 95% confidence.

---



**20.How are Z-scores used to compare different distributions?**
- Z-scores standardize different data points from various distributions, allowing for comparison on a common scale. By converting values to Z-scores, one can assess how far each value is from its respective mean in terms of standard deviations, facilitating comparisons across different datasets.

---



**21.What are the assumptions for applying the Central Limit Theorem?**
- The main assumptions for the Central Limit Theorem are:

 - The samples must be independent.
 - The sample size should be sufficiently large.
 - The population from which samples are drawn should have a finite mean and variance.

---



**22.What is the concept of expected value in a probability distribution?**
- The expected value (or mean) of a probability distribution is a measure of the center of the distribution, representing the long-term average outcome of a random variable. It is calculated as the sum of all possible values, each multiplied by its probability.

---





**23.How does a probability distribution relate to the expected outcome of a random variable?**
- A probability distribution describes how probabilities are assigned to different outcomes of a random variable. The expected outcome (expected value) is derived from this distribution, providing a weighted average of all possible outcomes based on their likelihoods.

**3. How do you write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?**

---



```python
import numpy as np

def bernoulli_pdf(x, p):
    """Calculate the PDF of a Bernoulli distribution."""
    return p if x == 1 else (1 - p) if x == 0 else 0

# Example usage
p = 0.3  # Probability of success
x_values = [0, 1]
pdf_values = [bernoulli_pdf(x, p) for x in x_values]

print("Bernoulli PDF values:", pdf_values)
```

**9.Write a Python function to calculate Z-scores from a dataset and plot them?**

---



```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """Calculate Z-scores for a given dataset."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 1000)  # Generate a normal distribution dataset

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plotting the Z-scores
plt.figure(figsize=(10, 6))
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Histogram of Z-scores')
plt.grid()
plt.show()
```

**10.How do you implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?**

---



```python
#The Central Limit Theorem states that the distribution of the sample means will
#approach a normal distribution as the sample size increases, regardless of the
#original distribution's shape. Below is a Python implementation that
#demonstrates this using a non-normal distribution
 #(specifically, an exponential distribution).
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 1000  # Number of samples
sample_size = 30    # Size of each sample
lambda_param = 1    # Rate parameter for the exponential distribution

# Generate samples from an exponential distribution
samples = np.random.exponential(scale=1/lambda_param, size=(num_samples, sample_size))

# Calculate the means of each sample
sample_means = np.mean(samples, axis=1)

# Plotting the original distribution and the distribution of sample means
plt.figure(figsize=(12, 6))

# Plot the original distribution
plt.subplot(1, 2, 1)
plt.hist(samples.flatten(), bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.title('Original Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
#Explanation:
#1.Parameters:
#1.num_samples: The number of samples to draw from the non-normal distribution.
#2.sample_size: The size of each sample.
#3.lambda_param: The rate parameter for the exponential distribution.
#2.Generate Samples:
#Samples are generated from an exponential distribution using
#NumPy's random.exponential function.
#3.Calculate Sample Means:
#The means of each sample are calculated using np.mean along the appropriate axis.
#4.Plotting:
#The first subplot shows the original exponential distribution.
#The second subplot shows the distribution of the sample means, which should
#approximate a normal distribution due to the Central Limit Theorem.'''
```

**14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**

---




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

# Example data
data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-score for a specific data point
data_point = 1.5
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for {data_point}: {z_score:.2f}")

# Plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, color='blue')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.title('Standard Normal Distribution with Z-score')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

**15.Implement hypothesis testing using Z-statistics for a sample dataset.**

---



```python
import numpy as np
from scipy import stats

# Sample data
sample_data = np.random.normal(100, 15, 30)  # Sample of size 30
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
population_mean = 100  # Hypothesized population mean
alpha = 0.05  # Significance level

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(len(sample_data)))

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Decision
if p_value < alpha:
    print(f"Reject the null hypothesis (p-value: {p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis (p-value: {p_value:.4f})")
```

**16.Create a confidence interval for a dataset using Python and interpret the result**

---



```python
import numpy as np
import scipy.stats as stats

# Sample data
data = np.random.normal(100, 15, 30)  # Sample of size 30
confidence_level = 0.95

# Calculate mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)

# Calculate confidence interval
ci = stats.t.interval(confidence_level, len(data)-1, loc=mean, scale=std_err)
print(f"Confidence Interval: {ci}")
```

 **23.Implement Z-score normalization on a dataset and visualize its transformation.**

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset
data = np.random.normal(100, 15, 100)  # Sample of size 100

# Z-score normalization
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
normalized_data = (data - mean) / std_dev

# Visualization
plt.figure(figsize=(12, 6))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=20, color='orange', alpha=0.7, edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

---



```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """Calculate skewness and kurtosis of a dataset."""
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

# Generate a dataset
data = np.random.normal(100, 15, 100)  # Sample of size 100

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
